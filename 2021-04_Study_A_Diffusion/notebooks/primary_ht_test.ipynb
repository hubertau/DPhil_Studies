```python
import jsonlines
import unidecode

def iterator(jsonl_file):
    hts = []
    with jsonlines.open(jsonl_file,'r') as reader:
        for line in reader:
            tweets = line['data']
            for tweet in tweets:
                if 'entities' in tweet and 'hashtags' in tweet['entities']:
                    hts.append([unidecode.unidecode(i['tag'].lower()) for i in tweet['entities']['hashtags']])
                    # logging.info(hts)
    # hts = ' '.join(hts)
    hts = [item for sublist in hts for item in sublist]


    return hts

test_file = '../data/01_raw/01_group/timeline_171954269.jsonl'

iterator(test_file)
```
